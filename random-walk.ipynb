```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from multiprocessing import Pool
import random
```

```python
#Number of steps
N = 100
# Starting coordinate
A = (2021101078%5) + 1
print(A)
```

```python
def rand_choice():
    return np.random.choice([-1,1])
```

```python
def random_walk(A,N):
    return np.cumsum(np.random.choice([1, -1], size=N)) + A
#     pos = np.zeros(N)
    
#     pos[0] = A
#     for i in range(1,N):
#         pos[i] = pos[i-1] + rand_choice()
#     return pos
```

```python
pos = random_walk(A,N)
plt.plot(range(N),pos)
plt.xlabel("N")
plt.ylabel("Displacement")
plt.show()
```

```python
success_count = 0;

prob_list = []
walks = 100
trials = 10000

for i in range(1,walks):
    success_count = 0
    
    for j in range(trials):
        # Simulate the walk for A starting position and I Steps
        disp = random_walk(A,i);

        # Check if the final pos is the same as start
        last = disp[-1]
        if(last == 0):
            success_count += 1
    
    prob_list.append(success_count/trials)

print(prob_list)
plt.plot(range(walks-1), prob_list)
plt.xlabel("N");
plt.ylabel("Probabilities")
```

```python
B = -A
success_count = 0;

prob_list = []
walks = 100
trials = 10000

for i in range(1,walks):
    success_count = 0
    
    for j in range(trials):
        # Simulate the walk for A starting position and I Steps
        disp1 = random_walk(A,i);
        disp2 = random_walk(B,i)

        # Check if the final pos is same for both
        if(disp1[-1] == disp2[-1]):
            success_count += 1
    
    prob_list.append(success_count/trials)

print(prob_list)
plt.plot(range(walks-1), prob_list)
plt.xlabel("N");
plt.ylabel("Probabilities")
```
